```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
import pandas as pd
from pandas import Series, DataFrame
from patsy import dmatrices
from sklearn import datasets, svm
```

```python
df = pd.read_csv("data.csv")
df
```

```python
fig = plt.figure(figsize=(10, 4), dpi=1600)
plt.grid(which = 'major',linestyle = "dashed")
plt.scatter(df['2021'], df['Country'], color='red',label='2021', marker='o', s=100)
plt.scatter(df['2016'], df['Country'], color='orange',label='2016', marker='o', s=100)
plt.scatter(df['2011'], df['Country'], color='#FFCCCC',label='2011', marker='o', s=100)
plt.legend(loc='best')
plt.show()

```

```python
fig, ax = plt.subplots(figsize=(4, 5), dpi=1600)

for i in range(len(df)):
    ax.plot([df['2011'][i], df['2016'][i]], [df['Country'][i], df['Country'][i]], color='grey', linestyle='solid')
    ax.plot([df['2016'][i], df['2021'][i]], [df['Country'][i], df['Country'][i]], color='grey', linestyle='solid')

ax.scatter(df['2021'], df['Country'], color='red', label='2021', marker='o', s=30)
ax.scatter(df['2016'], df['Country'], color='orange', label='2016', marker='o', s=30)
ax.scatter(df['2011'], df['Country'], color='#FFCCCC', label='2011', marker='o', s=30)

ax.grid(which='major', linestyle='dashed')
ax.set_xlim(0,100)
ax.legend(loc='best')
plt.show()
```
